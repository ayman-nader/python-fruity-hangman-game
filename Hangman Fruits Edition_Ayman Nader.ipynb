## Description
> ##### The Hangman program randomly selects a secret word from a list of secret words.
> ##### The Game: Here, a random word (a fruit name) is picked up from our collection and the player gets limited chances to guess the fruit name correctly.
> ##### When a letter in that word is guessed correctly, that letter position in the word is made visible in the output.
> ##### In this way, all letters of the word are to be guessed before all the chances are over.
> ##### For convenience, the number of chances equals the word length + 2.
> ##### For example, word to be guessed is mango, then user gets 5 + 2 = 7 chances, as mango is a five letter word.
> ##### Make game repeats until player ends the game.
> ##### Bonus : Create a function that takes 3 parameters `Easy`, `Medium` and `Hard` where `Medium` has more difficult words to be guessed than `Easy` and so on.

```python

# importing the required tools, to avoid forgetting afterwards

import random

# line separator function for better readability and to avoid repetition
def print_line():
    "line separator"
    
    print("\n" + "-" * 30 + "\n")

# player chooses difficulty function > based on player choice > game picks a random secret fruit from the chosen list
def choose_difficulty() :
    "asks player for difficulty and returns a random hidden word."

    # secret words are fruits per instructions >>> seperated by difficulty levels

    easy = ["apple", "mango", "grape", "melon", "peach",
                "lemon", "berry", "plum", "kiwi", "pear",
                    "fig", "date", "lime", "guava", "olive"]

    medium = ["banana", "orange", "papaya", "cherry", "apricot",
                "avocado", "coconut", "lychee", "mandarin", "nectarine",
                    "persimmon", "tangerine", "cranberry", "mulberry", "blueberry"]

    hard = ["pomegranate", "watermelon", "dragonfruit",
                "passionfruit", "strawberry", "blackberry",
                    "raspberry", "jackfruit", "starfruit",
                        "cantaloupe", "honeydew", "grapefruit",
                            "elderberry", "gooseberry", "boysenberry"]
    
    # to avoid code repetition and errors, will use a while loop + try/except to validate player input

    while True:

        try:

            # for convenience, choices are limited to 1, 2 and 3 
            
            difficulty = int(input("Choose difficulty : (1) Easy , (2) Medium , or (3) Hard"))

            if difficulty == 1:
                secret_word = random.choice(easy)
                print_line()
                print("You chose Easy difficulty, lovely <3")
                return secret_word
            
            elif difficulty == 2:
                secret_word = random.choice(medium)
                print_line()
                print("You chose Medium difficulty, enjoy :)")
                return secret_word
            
            elif difficulty == 3:
                secret_word = random.choice(hard)
                print_line()
                print("You chose Hard difficulty, I like your bravery !!!")
                return secret_word
           
        except:
            print_line()
            print("Invalid input! Please enter 1, 2, or 3.")

# random messages function based on guesses > to make it more engaging:
def get_random_message(guess_is_correct):
    "returns a random message based on whether the player's guess is correct or not."

    correct_encourage = [
        "Correct! Bravo :)",
        "Great job! You're on fire :D",
        "Opaaa! That's right!",
        "Bravo, that one counts.",
        "Correct :D you're a hero for this stickman!",
        "Nice guess!",
        "Impressive! did someone help you?"]

    wrong_encourage = [
        "Nope! Wrong guess :(",
        "Oops! That's not it.",
        "Nope. Even the stickman is judging you...",
        "Wrong. Nice try though.",
        "Incorrect. You sure you want to save him?",
        "That's not it. Sorry stickman :(",
        "I like your confidence, but that's not correct."]
    
    if guess_is_correct:
        return random.choice(correct_encourage)
    else:
        return random.choice(wrong_encourage)

# welcome message function
def welcome_message():
    "prints welcome message"
    print("Welcome to Hangman (Fruits edition 🍉) made by Ayman Nader, an Epsilon AI student :)")

# game rules function
def game_rules():
    "prints game rules"
    print("Try to guess the secret fruit, one letter at a time.")
    print("You have a limited number of chances, so choose wisely to save this poor fakahani stickman!")
    print(
        """
         -----
         |   |
        \O/  |  < el7a2oniii !
         |   |
        / \  |
             |
      ------------
        """)

# game variables function
def game_variables():
    "initializes the game variables and returns them"
    secret_word = choose_difficulty()
    display = ["_"] * len(secret_word)
    guessed_letters = []

    # per instructions, chances = length of secret word + 2
    chances = len(secret_word) + 2

    print(f"You have {chances} chances.\nThe secret fruit has {len(secret_word)} letters.")
    print("Let's play. Good luck and have fun !")
    return secret_word, display, guessed_letters, chances


# main game function that contains the main loop and handles the game logic
def game_logic():

    print_line()
    game_rules()
    secret_word, display, guessed_letters, chances = game_variables()

    # main loop
    # game is on so long as there are chances remaining 
    # and the player hasn't guessed all the letters of the secret word

    while chances > 0 and "_" in display:

        # shows the player's progress and remaining chances

        print_line()
        print("Fruit:", ' '.join(display))
        print("Guessed letters:", ', '.join(guessed_letters))
        print(f"Remaining guesses: {chances}")

        # ask the player to guess

        guess = str(input("Guess a letter : ").lower())

        # validate player input

        if len(guess) != 1 or guess.isalpha() == False:

            print_line()
            print("Invalid input! Please enter one letter.")
            continue

        # handle repeated guesses

        if guess in guessed_letters:

            print_line()
            print("You've already guessed that letter :/ try a different one.")
            continue

        guessed_letters.append(guess)

        # handle correct guesses

        guess_is_correct = guess in secret_word

        if guess_is_correct:

            for i in range(len(secret_word)):

                if secret_word[i] == guess:
                    display[i] = guess

            print_line()
            print(get_random_message(True))

        # handle wrong guesses

        else:

            guess_is_correct = False

            chances -= 1

            print_line()
            print(get_random_message(False))

        # Win scenario

        if display.count("_") == 0:

            print(
                    """
                    -----
                    |   |
                        |  
                \O      |  < shokraaan <3
                 |\     |
                / \     |
               --------------
                """)
            print("You've won the game and saved him :D you're amazing!")
            print(f"The secret fruit was: '{secret_word}'")

        # Lose scenario

        elif chances == 0 and "_" in display :

            print(
                """
                -----
                |   |
                O   |  < el wadaa3 ya donia </3
               /|\  |
               / \  |
                    |
             -------------
                    """)
            print("You lost the game :'( ... We'll save him next time!")
            print(f"The secret fruit was: '{secret_word}'")


def play_fruity_hangman():
    "plays the game and asks for replay."
    
    welcome_message()
    
    # DRY principle > while loop to avoid code repetition

    while True:

        game_logic()

        #ask player if they want to play again

        again = input("Play again ? [Y/N]").lower()

        if again != "y":

            print_line()
            print("Thanks for playing <3 and have an amazing day!")
            break


# =========================
# START THE GAME
# =========================|

play_fruity_hangman() 
```
